## Exercice séries

### Partie 1

- Créer une table `series`
    - Avec une colonne d'id (avec le type spécifié ci-dessous) et bien la nommer `id`
        - `INTEGER PRIMARY KEY AUTOINCREMENT`
    - Le nom de la série : `name` ou `title`
        - `TEXT NOT NULL`
    - Le nom du réalisateur : `director`
        - `TEXT NOT NULL`
    - La date de sortie (date du premier épisode de la saison 1) : `air_date` ou `release_date`
        - Chercher comment stocker une date
        - Décider si elle peut être nulle ou non
    - La note de la série (IMDb / SensCritique / Allociné) : `rating`
        - `REAL` (chercher un type plus précis si besoin)
- Insérer les données de 2 à 5 séries (récupérables sur Wikipédia en une fois)
- Afficher toutes les lignes de la table

### Partie 2

À l'aide de recherches internet :

- Ajouter une colonne pour la date du dernier épisode (`ALTER TABLE` pour créer la colonne, puis `UPDATE` pour remplir les valeurs)
- Ajouter une colonne contenant le nombre de jours entre le premier et le dernier épisode (calculé en SQL, pas à la main)
