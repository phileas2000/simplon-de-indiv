# Vérifier qu'on peut se connecter à PostgreSQL localement

```python
import psycopg2


USER = "postgres"
PASSWORD = "COMPLETER" # écrire le mot de passe choisi lors de l'installation de PostgreSQL
PORT = 5432 # ou 5433 pour certains


conn = psycopg2.connect(
    user=USER,
    password=PASSWORD,
    host="localhost",
    port=PORT
)

# Create a cursor object
cur = conn.cursor()

# Execute a query
cur.execute("SELECT version();")

# Fetch and print the result of the query
db_version = cur.fetchone()
print(db_version)
```
