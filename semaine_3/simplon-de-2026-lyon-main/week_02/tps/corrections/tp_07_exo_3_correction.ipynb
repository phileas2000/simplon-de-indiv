Implémenter une classe fille Lyceen qui aura en plus un attribut classe et une méthode veillir() qui augmente son âge de 1
Implémenter une classe fille Enseignant qui aura en plus un attribut matière et une méthode noter_lyceen qui prend un argument une instance de classe Lycéen et affiche un message avec le nom de l'étudiant et une note aléatoire entre 0 et 20

```python
import random


class Personne:
    def __init__(self, nom, prenom, age):
        self.nom = nom
        self.prenom = prenom
        self.age = age


class Lyceen(Personne):
    def __init__(self, nom, prenom, age, classe):
        super().__init__(nom, prenom, age)
        self.classe = classe

    def vieillir(self):
        self.age += 1


class Enseignant(Personne):
    def __init__(self, nom, prenom, age, matiere):
        super().__init__(nom, prenom, age)
        self.matiere = matiere

    def noter_lyceen(self, lyceen):
        note = random.randint(0, 20)
        print(f"{lyceen.prenom} {lyceen.nom} obtient {note}/20 en {self.matiere}")


lyceen = Lyceen("Alderson", "Elliot", 17, "Terminale")
prof = Enseignant("Robot", "Angela", 45, "informatique")
prof.noter_lyceen(lyceen)
lyceen.vieillir()
print("Nouvel age :", lyceen.age)
```

```python
# Plus difficile : une mini classe facon BeautifulSoup
class MySoup:
    def __init__(self, html):
        self.html = html

    def find_all_p(self):
        contenus = []
        reste = self.html
        while "<p>" in reste and "</p>" in reste:
            debut = reste.index("<p>") + len("<p>")
            fin = reste.index("</p>")
            contenus.append(reste[debut:fin].strip())
            reste = reste[fin + len("</p>"):]
        return contenus

    def balises_equilibrees(self):
        pile = []
        i = 0
        while i < len(self.html):
            if self.html[i] == "<":
                fin = self.html.index(">", i)
                balise = self.html[i + 1:fin]
                if balise.startswith("/"):
                    if not pile or pile.pop() != balise[1:]:
                        return False
                elif not balise.endswith("/"):
                    pile.append(balise)
                i = fin + 1
            else:
                i += 1
        return not pile


html = "<div><p>Bonjour</p><p>Au revoir</p></div>"
soup = MySoup(html)
print(soup.find_all_p())
print("Balises equilibrees :", soup.balises_equilibrees())
```
