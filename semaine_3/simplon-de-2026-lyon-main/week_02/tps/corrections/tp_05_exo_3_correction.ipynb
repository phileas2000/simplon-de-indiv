## Exercice 3 - Le prénom au plus grand âge prédit (API Agify, sans clé)

L'API Agify prédit un âge moyen pour un prénom, sans clé. URL de base : `https://api.agify.io`, avec le paramètre `name`. L'âge se trouve dans `reponse.json()["age"]`.

- Pour chaque prénom de la liste fournie, récupérer l'âge prédit et construire un dictionnaire `{prénom: âge}`.
- Trouver le prénom dont l'âge prédit est le plus élevé.


```python
import requests

prenoms = ["jean", "lea", "marcel", "ines", "raymond"]


def age_predit(prenom):
    reponse = requests.get("https://api.agify.io", params={"name": prenom})
    return reponse.json()["age"]


ages = {}
for prenom in prenoms:
    ages[prenom] = age_predit(prenom)

plus_age = None
for prenom, age in ages.items():
    if plus_age is None or age > ages[plus_age]:
        plus_age = prenom

print(ages)
print("Prénom au plus grand âge prédit :", plus_age, "->", ages[plus_age], "ans")
```
