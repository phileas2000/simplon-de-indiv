## Exercice 2 - Météo de plusieurs villes (API Open-Meteo, sans clé)

L'API Open-Meteo donne la météo sans inscription ni clé. URL de base : `https://api.open-meteo.com/v1/forecast`, avec les paramètres `latitude`, `longitude` et `current_weather=true`. La température actuelle se trouve dans `reponse.json()["current_weather"]["temperature"]`.

- Faire une requête GET pour Lyon (latitude 45.75, longitude 4.85) et afficher sa température actuelle.
- Pour chaque ville du dictionnaire fourni, récupérer la température et construire un dictionnaire `{ville: température}`.


```python
import requests

villes = {
    "Lyon": (45.75, 4.85),
    "Paris": (48.85, 2.35),
    "Marseille": (43.30, 5.37),
}


def temperature(latitude, longitude):
    reponse = requests.get(
        "https://api.open-meteo.com/v1/forecast",
        params={"latitude": latitude, "longitude": longitude, "current_weather": True},
    )
    return reponse.json()["current_weather"]["temperature"]


print("Température à Lyon :", temperature(45.75, 4.85))

temperatures = {}
for ville, (latitude, longitude) in villes.items():
    temperatures[ville] = temperature(latitude, longitude)

print(temperatures)
```
