### Standard deviation
$$ \sqrt{ \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x} ) ^2 } $$

## Maximum

```python
# Essaie naif
def get_maximum(l):
    current_max = l[0]
    # current_max c'est le max de toutes les valeurs déjà vues
    for value in l:
        if value > current_max:
            current_max = value
    return current_max

max_ = get_maximum(liste_notes)
max_


# Un peu mieux
def get_maximum(l):
    if not l:
        raise ValueError("List l is empty")
    current_max = l[0]
    # current_max c'est le max de toutes les valeurs déjà vues
    for k in range(1, len(l)):
        value = l[k]
        if value > current_max:
            current_max = value
    return current_max

#max_ = get_maximum(liste_notes)
#max_ = get_maximum([])
```

## Moyenne

```python
import math
math.sqrt(2)
```

```python
def get_mean(l):
    if not l:
        raise ValueError("List l is empty")
    
    sum_ = 0
    for value in l:
        sum_ += value
        
    return sum_ / len(l)


# get_mean([])
#get_mean(liste_notes)

def get_mean(l):
    if not l:
        raise ValueError("List l is empty")
    return sum(l) / len(l)

get_mean(liste_notes)


def get_standard_deviation(l):
    if not l:
        raise ValueError("List l is empty")    
    mean = get_mean(l)
    return math.sqrt(sum([(x-mean) **2 for x in l]) / len(l))

get_standard_deviation(liste_notes)
    
    
    
def get_median(l):
    l.sort()
    # Longueur impaire
    if len(l) % 2 == 1:
        index = (len(l) + 1) / 2
        median = l[index]
        
    else:
        #      o  ml mr o
        # l = [1, 2, 3, 4]
        index_mid_right = int(len(l) / 2)
        index_mid_left = index_mid_right - 1
        median = (l[index_mid_right] + l[index_mid_left]) / 2
        
    return median
            
```

```python
index_mid_right = int(len(liste_notes) / 2)
index_mid_left = index_mid_right - 1
print(liste_notes[index_mid_left], liste_notes[index_mid_right])
```
