## Exercice 4 - Compter les lignes et les mots d'un fichier

On écrit d'abord un petit texte dans un fichier (donnée fournie), puis tu le relis.

- Écrire une fonction `compter(chemin)` qui ouvre le fichier et renvoie un couple `(nombre_de_lignes, nombre_de_mots)`.
- L'appliquer au fichier `data/tp_04/poeme.txt`.


```python
# Sorties ecrites dans corrections/data/ (executer ce notebook depuis le dossier tps/)
# Donnée fournie : on crée le fichier à analyser
texte = """Python est un langage clair.
On apprend a lire et a ecrire des fichiers.
Un fichier est une suite de lignes."""
with open("corrections/data/tp_04/poeme.txt", "w") as f:
    f.write(texte)

# Votre code ici
```

```python
# Sorties ecrites dans corrections/data/ (executer ce notebook depuis le dossier tps/)
def compter(chemin):
    with open(chemin) as f:
        contenu = f.read()
    nombre_de_lignes = len(contenu.split("\n"))
    nombre_de_mots = len(contenu.split())
    return nombre_de_lignes, nombre_de_mots


print(compter("corrections/data/tp_04/poeme.txt"))
```
