## Exercice 5 - Fusionner plusieurs fichiers

On crée d'abord trois fichiers (donnée fournie).

- Écrire une fonction `fusionner()` qui lit tous les fichiers dont le nom commence par `partie_` dans `data/tp_04/`, et écrit leur contenu mis bout à bout dans `data/tp_04/fusion.txt`.
- Indication : `os.listdir("data/tp_04")` donne la liste des noms de fichiers du dossier.


```python
# Sorties ecrites dans corrections/data/ (executer ce notebook depuis le dossier tps/)
import os

# Donnée fournie : on crée trois fichiers à fusionner
for i in range(1, 4):
    with open(f"corrections/data/tp_04/partie_{i}.txt", "w") as f:
        f.write(f"Contenu de la partie {i}\n")

# Votre code ici
```

```python
# Sorties ecrites dans corrections/data/ (executer ce notebook depuis le dossier tps/)
import os


def fusionner():
    fichiers = sorted(nom for nom in os.listdir("data/tp_04") if nom.startswith("partie_"))
    with open("corrections/data/tp_04/fusion.txt", "w") as sortie:
        for nom in fichiers:
            with open(f"corrections/data/tp_04/{nom}") as f:
                sortie.write(f.read())
    with open("corrections/data/tp_04/fusion.txt") as f:
        print(f.read())


fusionner()
```
