## Exercice 1 : Leet Code Problem : Two Sums

Le problème des "Two Sum" est un problème algorithmique classique souvent utilisé lors des entretiens de programmation. 

Étant donné : 

- un tableau d'entiers `nums` 
- un entier `target` 

Renvoyez les indices des deux nombres tels que leur somme vaut `target`. 
Vous pouvez supposer que chaque entrée aura exactement une solution et vous ne pouvez pas utiliser le même élément deux fois.


```python
def two_sum(numbers_list,target):
    for i1,num_1 in enumerate(numbers_list):
        print(f"i1: {i1} num_1: {num_1}")
        for i2,num_2 in enumerate(numbers_list):
            print(f"i2: {i2} num_2: {num_2}")
            if num_1+num_2==target and i1!=i2:
                return i1,i2
two_sum([2,4,5,3,8],8)
```

```python
def tournament_round(l,if_counter):
    winner_list=[]
    loser_list=[]
    if_counter[0]+=1
    if len(l)%2!=0:
        e=l.pop(random.randrange(len(l)))
        winner_list.append(e)
    for _ in range(len(l)//2):
        e1=l.pop(random.randrange(len(l)))
        e2=l.pop(random.randrange(len(l)))
        if_counter[0]+=1
        if e1>e2:
            winner_list.append(e1)
            loser_list.append(e2)
        else:
            winner_list.append(e2)
            loser_list.append(e1)
    return winner_list,loser_list
```

```python
def finish_sorting(l,if_counter,silent=False):
    max_rank=len(l)-1
    ended=False
    while ended==False:
        ended=True
        for current_rank in range(max_rank):
            if_counter[0]+=1
            if l[current_rank]<l[current_rank+1]:
                buffer=l[current_rank]
                l[current_rank]=l[current_rank+1]
                l[current_rank+1]=buffer
                ended=False
        max_rank-=1
    if silent==False:
        print("Final if_counter: "+str(if_counter))
        print("silent==False "+str(silent==False))
    return l
```

```python
def tournament_sort(master_list,silent=False):
    if_counter=[0]
    og_len=len(master_list)
    
    master_list=[master_list]
    if silent==False:
        print("og len "+str(og_len))
    ended=False
    while ended==False:
        ended=True
        if silent==False:
            print("Prefiltering if_counter: "+str(if_counter))
            print(master_list)     
        new_master_list=[]
        for i in range(len(master_list)):
            if_counter[0]+=1
            if len(master_list[i])<3:
                new_master_list.append(master_list[i])
            else:
                winner_list,loser_list=tournament_round(master_list[i],if_counter)
                new_master_list.append(winner_list)
                new_master_list.append(loser_list)
                ended=False
        master_list=new_master_list
    if silent==False:
        print("Intermediary if_counter: "+str(if_counter))
    semi_sorted=list(deepflatten(master_list))
    if silent==False:
        print(semi_sorted)
    sorted=finish_sorting(semi_sorted,if_counter,True)

    return sorted,if_counter
```
